```python
#imports
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
#load csv file
sfo_data_df=pd.read_csv(Path("../PyViz_homework06/sfo_neighborhoods_census_data.csv"))
```

```python
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

```python
housing_units_by_year=sfo_data_df.groupby('year').mean()
housing_units_by_year
```

```python
housing_units_by_year.hvplot.bar(x="year",y="housing_units",color="blue",xlabel="Year",ylabel="Housing Units",
                                 label="Housing Units in San Francisco from 2010 to 2016" ).opts(yformatter="%0f")
```

```python
#What's the overall trend?
#Clearly there is an upward trend in housing units during the period of 2010-2016 in San Francisco
```

```python
#group by year and take average
sfo_data_df.groupby('year').mean()

#Table below shows lowest gross rent between 2010-2016 is $1239.00 (occurred in 2010)
```

```python
#New dataframe
prices_square_foot_by_year=sfo_data_df.groupby("year").mean()
```

```python
#drop column "housing units"
prices_square_foot_by_year=prices_square_foot_by_year[["sale_price_sqr_foot","gross_rent"]]
prices_square_foot_by_year.head()
```

```python
#Line plot 
prices_square_foot_by_year.hvplot.line(
title="Sale Price per Square Foot and Average Gross Rent- 2010-2016 - San Francisco",
xlabel="Year",
ylabel="Gross Rent/Sale Price per Square Foot"
)
```

```python
#Analysis
#2011 was the only year that experienced a drop in average sale price per square foot but the gross rent increased regardless
```

```python
#New df grouping year and neighborhood by mean
prices_by_year_by_neighborhood=sfo_data_df.groupby(["year","neighborhood"]).mean()
```

```python
#Remove "housing_units" column
prices_by_year_by_neighborhood=prices_by_year_by_neighborhood[["sale_price_sqr_foot","gross_rent"]]
```

```python
display(prices_by_year_by_neighborhood.head())
display(prices_by_year_by_neighborhood.tail())
```

```python
#Create line plot
prices_by_year_by_neighborhood.hvplot.line(
                                           x="year",
                                           xlabel="Year",
                                           ylabel="Gross Rent / Sale Price per Square Foot",
                                           groupby="neighborhood",
                                           title="Sale Price per Square Foot and Average Gross Rent by Neighborhood (2010-2016)")
```

```python
#Answer question- The sale price per square foot in Anza Vista in 2016 is less than the price in 2010; $88.40 vs $267.93 respectively 
```

```python
#Build interactive neighborhood map, create new df
neighborhood_locations_df=pd.read_csv(Path("../PyViz_homework06/neighborhoods_coordinates.csv"),index_col="Neighborhood")
```

```python
neighborhood_locations_df.head()
```

```python
#Create new df that groups data by neighborhood with mean
all_neighborhood_info_df=sfo_data_df.groupby("neighborhood").mean()
```

```python
#Concatenate neighborhood_locations and all_neighborhood_info on column axis
all_neighborhoods_df=pd.concat([neighborhood_locations_df,all_neighborhood_info_df],axis="columns",sort=False)
#Display new df
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
#Remove null values
all_neighborhoods_df=all_neighborhoods_df.reset_index().dropna()
#Rename index column to "Neighborhood"
all_neighborhoods_df=all_neighborhoods_df.rename(columns={"index":"Neighborhood"})
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
#Create points plot
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    color='gross_rent',
    size='sale_price_sqr_foot',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    title="Average Sale Price per Square Foot and Gross Rent in San Francisco")
    
```

```python
#QUESTION - Which neighborhood has the highest gross rent and which has the highest sale price per square foot?
#Union Square has the highest sale price per square foot $903.99 and Westwood Park has the highest gross rent $3959
```

##Summary
Gross rent has increased every year during the 2010-2016 period.  Sale price per square foot has risen every year with the exception of 2010-2011. 
The growth in rental income has far outpaced the growth in sales price during 2010-2016.  This is great for a buy-and-rent strategy.  Some neighborhoods in the data set have experienced downwad and flat sales prices with increasing rents making them particularly attractive to investors.
